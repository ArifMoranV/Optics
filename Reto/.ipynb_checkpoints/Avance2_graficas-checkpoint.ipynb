$$\begin{bmatrix}
r_f\\
\theta_f
\end{bmatrix}=T \begin{bmatrix}
r_i\\
\theta_i
\end{bmatrix}$$


$$M=-\frac{S_i}{S_o}$$



$$M=\frac{f_1}{f_2} ; f_1=M f_2 \therefore  f_1=\frac{S_i S_{i2}}{S_o S_{o2}} f_2  $$


\begin{align}\begin{bmatrix}
r_f\\
\theta_f
\end{bmatrix}=\begin{bmatrix}
1 & d_2\\
0 & 1
\end{bmatrix}\begin{bmatrix}
1 & 0\\
-\frac{1}{f_2} & 1
\end{bmatrix}\begin{bmatrix}
1 & d\\
0 & 1
\end{bmatrix}\begin{bmatrix}
1 & 0\\
-\frac{1}{f_1} &1
\end{bmatrix}\begin{bmatrix}
1 & d_1\\
0 & 1
\end{bmatrix}\begin{bmatrix}
r_i\\
\theta_i
\end{bmatrix}
\end{align}



```julia
N=6 ###Numero de rayos
f_1=0.54;
f_2=0.006;
d_1=7#384400*1000;
d_2=3#abs(f_2);
d=1
θ=0*pi/4
θi=ones(N)*θ
θi[Int(N/2)+1:end]*=-1    #0*pi/4;###Angulo de indicencia

yo=10
```

```julia

r_is=range(-yo,yo,N)
aux=ones(N)*(d_1)*2
xs=zeros(N);
gr(size=(1000,800))
scatter(xs,r_is,label="")
plot!(xs.+d_1,r_is.*2,color="blue",lw=2,label=L"f_1")### Lente 
plot!(xs.+(d_1+d),r_is.*2,color="red",lw=2,label=L"f_2")
quiver!(xs,r_is,quiver=(aux.*cos.(θi),aux.*sin.(θi)),color="black",label=L"f_2")
X=range(0,(d_1+d_2+d)*1.3,10)
plot!(X,X.*0,color="orange",label="Zona paraxial")

```

```julia
function Mati(input_vector)
   output_matrix = zeros(size(input_vector)[1], 6)
    for i in 1:size(input_vector)[1]
        input_tuple = input_vector[i]
         
        # Unpack the tuple into three arrays
        array1, array2, array3 = input_tuple
        
        # Fill the corresponding rows of the output matrix
        output_matrix[i, 1:2] = array1
        output_matrix[i, 3:4] = array2
        output_matrix[i, 5:6] = array3
    end
    return output_matrix
end
```

```julia

function Cf(d_1,d_2,f_1,f_2,ri,θi,d)
    function PLL()
        Ci=[ri;θi]
        Pd1=[1 d_1;0 1]##### Propagacion
        L1=[1 0;(-1/f_1) 1]
        ######Lente 1
        PL=L1*Pd1*Ci
    return Ci,Pd1,L1,PL
    end
    ####### lENTE 2
    function PLL2()
        P=[1 d;0 1]###########################################
        L2=[1 0;(-1/f_2) 1]
        #PL2=L2*P
        return P,L2
    end

    Ci,Pd1,L1,PL=PLL()
    P,L2=PLL2()
    Pd2=[1 d_2; 0 1]
    T=Pd2*L2*P*L1*Pd1*Ci
    return T,PL,L2*P*L1*Pd1*Ci
end
```

```julia
NN=100
θs=range(0,4*2*pi,NN);
rs=range(0,100,NN);
rfs=ones(NN);
θfs=ones(NN);
```

```julia
Cf(d_1,d_2,f_1,f_2,ri,θi)
```

```julia
for i in range(1,NN)
    rfs[i],θfs[i]=Cf(d_1,d_2,f_1,f_2,ri,θs[i])
end
```
